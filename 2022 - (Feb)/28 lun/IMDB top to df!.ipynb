###  Build iterator, create list urls

```python
iterations = range(1,2283,50)
```

```python
for i in iterations:
    start_at=str(i)
    url='https://www.imdb.com/search/title/?title_type=feature&release_date=1990-01-01,1995-01-01&user_rating=7.0,&start='+start_at+'&ref_=adv_nxt'
    print(url)
```

### respectful scraping - use sleep and randint to separate scraper pings

```python
# example to simulate human behaviour
for i in range(5):
    print(i)
    sleep (3)
```

```python
# example 2, more human: randomly sleep
for i in range(5):
    print(i)
    wait_time=randint(1,4)
    print('I will sleep now for '+str(wait_time)+' seconds')
    sleep (wait_time)
```

### Assemble the scraper

```python
pages=[]
for i in iterations:
    start_at=str(i)
    url='https://www.imdb.com/search/title/?title_type=feature&release_date=1990-01-01,1995-01-01&user_rating=7.0,&start='+start_at+'&ref_=adv_prv'
    response=requests.get(url)
    #just for monitoring 
    print("status="+str(response.status_code))
    pages.append(response)
    # take a short sleep 
    wait_time=randint(1,4)
    print("I will sleep now for "+str(wait_time)+' seconds')
    sleep(wait_time)  
```

```python
# create soup from one page

soup=BeautifulSoup(pages[0].content, 'html.parser')
```

### Final iterator
* each page response
* then parse each page
* capture relevant text
* then for each parse get title and synopsis
* clean as needed, get into dataframe

```python
pages_parsed = []
synopsis = []
title = []
for i in tqdm(range(len(pages))):
    pages_parsed.append(BeautifulSoup(pages[i].content,'html.parser'))
    movies_html=pages_parsed[i].select('div.lister-item-content')
    for a in range(len(movies_html)):
        title.append(movies_html[a].select('h3 > a')[0].get_text())
        synopsis.append(movies_html[a].select("p:nth-child(4)")[0].text[1:][:-1])
```

```python
all_movies = pd.DataFrame({'Movie': title, 'Synopsis': synopsis})
```
